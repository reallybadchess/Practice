### What were the most popular female and male baby names for each year?

```python
# slicing out the rows by state
ca_babies = states_babies.loc[states_babies['State'] == 'CA', :]
ca_babies.head()
```

```python
def popular(s):
    '''
    Receives s, a Pandas Series, containing baby names in 
    order of highest count to lowest count.
    Returns the most popular name in s.
    '''
    return s.iloc[0]
```

```python
# grouping by year and gender
ca_grouped = ca_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)
```

```python
ca_grouped
```

## Exploratory Data Analysis (EDA)

```python
# Importinng crime file

crime = pd.read_csv('crime_boston.csv')
crime.head()
```

```python
crime.groupby('INCIDENT_NUMBER').count()
```

## Data Cleaning

```python
# looking for missing values with .isnull() method
crime.isnull().any(axis=1)
```

```python
# finding the rows with missing values
rows_miss_vals = crime.isnull().any(axis=1)
crime[rows_miss_vals]
```

```python
crime.info()
```

```python
# dropping rows with missing vals = NaN
clean_crime = crime.dropna(how='all')
clean_crime.head()
```

```python
# we want to drop month, year, and date
clean_crime = clean_crime.drop(columns =['YEAR','MONTH','HOUR'])
clean_crime.head()
```

```python
clean_crime['OFFENSE_CODE_GROUP'].unique()
```

```python
clean_crime['OFFENSE_DESCRIPTION'].unique()
```

```python
clean_crime['DAY_OF_WEEK'].unique()
```

```python
# dropping location column
clean_crime = clean_crime.drop(columns = 'Location')
clean_crime
```

## Data Visualization

```python
# making a count plot
sns.set(rc={"figure.figsize":(15, 6)}) #width=10, height=6
sns.countplot(x ='DISTRICT', data=clean_crime)
plt.xticks(rotation = 90)
plt.show()
```

```python
# barplot
sns.barplot(x = 'STREET', data = clean_crime)
plt.xticks(rotation = 90)
plt.show()
```
